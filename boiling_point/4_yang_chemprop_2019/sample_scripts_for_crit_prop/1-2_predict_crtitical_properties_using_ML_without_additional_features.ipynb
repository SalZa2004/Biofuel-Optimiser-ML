# Make predictions using the ML model that is trained without any additional features

Before you run this script, make sure to:
- Use chemprop environment
- Download the `CritProp_ML_model_files_without_additional_feat.zip` file from https://zenodo.org/record/8072892, unzip it, and move it under `chemprop/crit_prop_model_files`


```python
import os
import chemprop
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

## Make predictions

```python
# Set the path to the model file and arguments

curr_dir = os.getcwd()
model_dir = os.path.join(curr_dir, '../crit_prop_model_files/CritProp_ML_model_files_without_additional_feat')


arguments = [
    '--test_path', 'crit_prop_sample_dataset.csv',
    '--preds_path', 'crit_prop_preds_without_additional_features.csv',
    '--checkpoint_dir', model_dir
]
```

```python
# make predictions

args = chemprop.args.PredictArgs().parse_args(arguments)
preds = chemprop.train.make_predictions(args=args)
```

## Compare the prediction results with the experimental data

The model is trained on these data, so the predictions should give very good agreement

```python
df_data = pd.read_csv('crit_prop_sample_dataset.csv')
df_pred = pd.read_csv('crit_prop_preds_without_additional_features.csv')
```

```python
# plot the best models separately

chosen_target_type_list = ['Tc', 'Pc', 'rhoc', 'omega']
unit_dict = {'Tc': 'K', 'Pc': 'bar', 'rhoc': 'mol/L', 'omega': '-'}

fig = plt.figure(figsize=(22, 4))

# Plot spacing
wspace = 0.35

plot_number = 1
for target_type in chosen_target_type_list:
    # plot the subplot
    ax = fig.add_subplot(1, 4, plot_number)
    plt.subplots_adjust(wspace=wspace)
    true_list = df_data[target_type].to_list()
    pred_list = df_pred[target_type].to_list()
    plt.plot(true_list, pred_list, '.', markersize=10, linestyle = 'None', alpha=0.5,
             color='#C0462E')
    plt.title(target_type, fontsize=24)
    ax.set_xlabel(f'Expt {target_type} ({unit_dict[target_type]})', fontsize = 20)
    ax.set_ylabel(f'Pred {target_type} ({unit_dict[target_type]})', fontsize = 20)
    axmin = min(min(true_list), min(pred_list)) - 0.1*(max(true_list)-min(pred_list))
    axmax = max(max(true_list), max(pred_list)) + 0.1*(max(true_list)-min(pred_list))
    ax.set_xlim(axmin, axmax)
    ax.set_ylim(axmin, axmax)

    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)


    # Display errors
    R2 = '%.2f' % r2_score(true_list, pred_list)
    mae = mean_absolute_error(true_list, pred_list)
    rmse = mean_squared_error(true_list, pred_list, squared=False)

    ax = fig.add_subplot(1, 4, plot_number, frame_on=False)   # creating a single axes
    plt.subplots_adjust(wspace=wspace) # this line sets the spacing between each plot
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)

    ax.text(0.05, 0.9, r'$R^2$: ' + R2, fontsize=20)
    ax.text(0.05, 0.8, 'mae: ' + "%.2f" % mae, fontsize=20)
    ax.text(0.05, 0.7, 'rmse: ' + "%.2f" % rmse, fontsize=20)

    plot_number += 1
        
```
